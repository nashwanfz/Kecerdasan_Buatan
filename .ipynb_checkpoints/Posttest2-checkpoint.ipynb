```python
# Import dulu semua library yang kepake
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Biar grafiknya lebih cakep
sns.set(style="whitegrid")
```

```python
# Baca dataset
df = pd.read_csv("SuperStoreOrders.csv")

# Liat 5 baris teratas buat ngecek datanya
df.head()
```

```python
# Bagian 1
# Disini kita mau ngeliat statistik dasar dari kolom angka
# Kayak mean, std, min, max, plus Q1, Q2 (median), Q3
stats = df.select_dtypes(include='number').describe().T

Q1 = df.select_dtypes(include='number').quantile(0.25)
Q2 = df.select_dtypes(include='number').quantile(0.50)
Q3 = df.select_dtypes(include='number').quantile(0.75)

stats['Q1'] = Q1
stats['Q2'] = Q2
stats['Q3'] = Q3

stats
```

```python
# Bagian 2
# Nah ini kita ambil 10 data awal khusus yang tipe angka aja
df.select_dtypes(include='number').head(10)
```

```python
# Bagian 3
# Kita bikin ringkasan dataset:
# - jumlah baris
# - jumlah kolom
# - jumlah kolom numerik
# - nama-nama kolom
info_dataset = pd.DataFrame({
    "Jumlah Record": [df.shape[0]],
    "Jumlah Attribute": [df.shape[1]],
    "Jumlah Attribute Numerik": [df.select_dtypes(include='number').shape[1]],
    "Nama Attribute": [list(df.columns)]
})

info_dataset
```

```python
# Bagian 4
# Cek berapa banyak kategori unik dari tiap kolom yang tipe object
object_cols = df.select_dtypes(include='object')

label_counts = pd.DataFrame({
    "Attribute": object_cols.columns,
    "Jumlah Label Unik": [df[col].nunique() for col in object_cols.columns]
})

label_counts
```

```python
# Bagian 5
#kita liat distribusi sales pake histogram + KDE biar keliatan polanya
plt.figure(figsize=(8,5))
sns.histplot(df['sales'], bins=50, kde=True)
plt.title("Distribusi Penjualan (Sales)")
plt.xlabel("Sales")
plt.ylabel("Frekuensi")
plt.show()
```

```python
# Bagian 6
# Bikin heatmap buat liat hubungan antar kolom numerik
plt.figure(figsize=(8,6))
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Heatmap Korelasi Atribut Numerik")
plt.show()
```

Kesimpulan
- Distribusi kolom Sales condong ke kanan atau right skewed artinya mayoritas penjualan bernilai kecil sementara ada beberapa transaksi dengan nilai sangat besar  
- Kolom Profit memiliki variasi besar terdapat transaksi yang merugi bernilai negatif dan ada juga yang untung besar bernilai positif  
- Hasil korelasi menunjukkan  
  - Kolom Sales cukup berhubungan dengan kolom Profit  
  - Kolom Discount berkorelasi negatif dengan kolom Profit semakin besar diskon maka profit cenderung menurun  
  - Kolom Quantity dan kolom Shipping Cost tidak memiliki pengaruh yang kuat terhadap kolom Profit  
- Sebagian besar transaksi dilakukan dengan jumlah barang yang relatif kecil biasanya hanya beberapa unit saja
